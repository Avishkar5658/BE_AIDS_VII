```python
# Step No.4 - Group sales data by region and calculate total sales amount.
sales_by_region = df.groupby('Sales_Region')['Sales_Amount'].sum().reset_index()
sales_by_region
```

Step No.5 - Create bar plots or pie charts to visualize the sales distribution by region

```python
# Bar plot
plt.figure(figsize=(10, 6))
plt.bar(sales_by_region['Sales_Region'], sales_by_region['Sales_Amount'])
plt.xlabel('Sales Region')
plt.ylabel('Total Sales Amount')
plt.title('Sales Distribution by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(sales_by_region['Sales_Amount'],labels=sales_by_region['Sales_Region'],
autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.show()
```
