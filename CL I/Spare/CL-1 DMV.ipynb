```python
df.columns = [col.lower() for col in df.columns]
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['unit price'] = df['unit price'].astype(float)
```

```python
df['quantity'] = df['quantity'].astype(int)
```

```python
df['total'] = df['unit price'] * df['quantity']
df['total'] = df['total'].astype(float)
```

```python
df.info()
```

```python
df['total sales'] = df['unit price'] * df['quantity']
df
```

```python
total_sales = df['total sales']  
print(total_sales.sum())
```

```python
average_order_value = df['total sales'].mean()
average_order_value
```

```python
category_sales = df.groupby('product line')['total sales'].sum()
category_sales
```

```python
import matplotlib.pyplot as plt
```

```python
category_sales.plot(kind='bar', title='Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
category_distribution = df['product line'].value_counts()
category_distribution.plot(kind='pie',title='Product Category Distribution',
autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
df.boxplot(column='total sales',by='product line', grid=False, rot=45)
plt.title('Sales Distribution by Product Category')
plt.suptitle('')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```
